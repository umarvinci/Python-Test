```python
# function for moving average
def moving_average(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
# moving average for df_TCS_1
windows = list(range(4,53,12))
d_TCS_MA=dict()
for window in windows:
    a=df_TCS_1.iloc[:,0].values
    d_TCS_MA[window] = moving_average(a,window*7)
    
```

```python
#print(len(df_TCS_MA[16]))
d_TCS_MA[16]
```

```python
# moving average for df_INFY_1 
windows = list(range(4,53,12))
d_INFY_MA=dict()
for window in windows:
    #a=df_TCS_1.iloc[:,0].values
    d_INFY_MA[window] = moving_average(df_INFY_1.iloc[:,0].values ,window*7)
  
```

```python
# moving average for df_niftt_1
windows = list(range(4,53,12))
d_niftt_1_MA=dict()
for window in windows:
    #a=df_TCS_1.iloc[:,0].values
    d_niftt_1_MA[window] = moving_average(df_niftt_1.iloc[:,0].values ,window*7)
  
```

```python
len(Low)
d={'Open':Open,'High':High,'Low':Low,'Close':Close,'Volume':Volume,'TurnOver':TurnOver}
for k in d:
    print(len(d[k]))
```

```python
df_dummy_TS = pd.DataFrame({'Open':Open,'High':High,'Low':Low,'Close':Close,'Volume':Volume,'TurnOver':TurnOver})
df_dummy_TS.head()
```

```python
date4dummy = pd.date_range(pd.datetime.today(), periods=1000)#.tolist()
```

```python
df_dummy_TS.index = pd.Series(date4dummy.date)#.apply(lambda x:x[:12])
```

```python
df_dummy_TS.head()
```

```python
df_dummy_TS ['Volume_pct_change'] = df_dummy_TS['Volume'].pct_change()*100
df_dummy_TS['Close_pct_chage'] = df_dummy_TS['Close'].pct_change(-1)*100
df_dummy_TS.head()
```

```python
def give_Volume_shock(x):
    if type(x)== float:
        if x > 10:
            return 1
        elif x < -10:
            return 1
        else:
            return 0
    else:
        return np.nan
def give_Price_shock(x):
    if type(x)== float:
        if x > 2:
            return 1
        elif x < -2:
            return 1
        else:
            return 0
    else:
        return np.nan
```

```python
df_dummy_TS['Price_shock']=df_dummy_TS['Close_pct_chage'].apply(give_Price_shock)
df_dummy_TS['Volume_shock']=df_dummy_TS['Volume_pct_change'].apply(give_Volume_shock)
```

```python
df_dummy_TS.head()
```

```python
def give_P_shock_No_V_shock(df):
    l=[]
    for i in range(len(df)):
        if df.iloc[i,0]==1 and df.iloc[i,1]==0:
            #print('yes')
            l.append(1)
        else:
            l.append(0)
    return l
```

```python
#give_P_shock_No_V_shock(df_dummy_TS.iloc[:,-2:])
a=df_dummy_TS.loc[:,['Price_shock','Volume_shock']]
#give_P_shock_No_V_shock(a)
```

```python
#p_yes_V_no = give_P_shock_No_V_shock(a)
df_dummy_TS['P_shock_yes_V_shoc_No'] = give_P_shock_No_V_shock(a)
df_dummy_TS.head(10)
```
